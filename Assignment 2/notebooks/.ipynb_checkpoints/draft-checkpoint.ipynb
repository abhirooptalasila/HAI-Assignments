```python
%matplotlib inline
import sys
sys.path.append("..")

import os
import gzip
import torch 
import numpy as np
from tqdm import tqdm
from skimage import io
from pathlib import Path
import albumentations as A
import matplotlib.pyplot as plt
from torchsummary import summary
from torchvision import transforms
from skimage.transform import resize
from PIL import Image, ImageSequence
from torch.utils.data import DataLoader
from os.path import join, abspath as abs
from torch.utils.data import random_split
from albumentations.pytorch.transforms import ToTensorV2

from unet.datasets import ISBI, DS_Bowl
from unet.utils import process_ISBI, visualize, soft_dice_loss
from unet.models import UNet

ISBI_ = abs("../data/isbi")
BOWL_ = abs("../data/nuclei")
BRATS_ = abs("../data/brats13")
LITS_ = abs("../data/lits")
IMG_H, IMG_W = 256, 256
```

```python
# just required for visualization
imgs_1 = io.imread(join(ISBI_, "train-volume.tif"))
masks_1 = io.imread(join(ISBI_, "train-labels.tif"))

print(imgs_1.shape, masks_1.shape)
visualize(imgs_1[0], masks_1[0])

# process ISBI .tiff files. Run only once.
# process_ISBI(imgs_1, masks_1, ISBI_)
```

```python
iaugs = A.Compose([
    A.Resize(IMG_H, IMG_W), 
    #A.Normalize(mean = [0.5,0.5,0.5], std = [0.5,0.5,0.5])
], p=1)

maugs = A.Compose([
    A.Resize(IMG_H, IMG_W, interpolation=Image.NEAREST)
], p=1)

bowl_ds = DS_Bowl(BOWL_, iaugs, maugs)
img, mask = bowl_ds.__getitem__(0)
print(img.shape, mask.shape)

length = bowl_ds.__len__()
splits = [int(length - 0.2*length), int(0.2*length)] 
bowl_train, bowl_val = random_split(bowl_ds, splits, generator=torch.Generator().manual_seed(42))

bowl_tdl = DataLoader(bowl_train, batch_size=4, shuffle=True)
bowl_vdl = DataLoader(bowl_val, batch_size=4, shuffle=True)

train_features, train_labels = next(iter(bowl_tdl))
val_features, val_labels = next(iter(bowl_vdl))
print("Training batch shape:", train_features.size(), train_labels.size())
print("Validation batch shape:", val_features.size(), val_labels.size())
```

```python
model = UNet(3, 1)

device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")

if device.type == "cuda":
    model.cuda()
    model = model.type(torch.cuda.FloatTensor)

criterion = torch.nn.BCEWithLogitsLoss()
optimizer = torch.optim.Adam(model.parameters(), lr = 1e-3)
epochs = 1
min_valid_loss = np.inf
 
for e in range(epochs):
    train_loss = 0.0
    for data, labels in tqdm(bowl_tdl):
        if device.type == "cuda":
            data = data.type(torch.cuda.FloatTensor)
            labels = labels.type(torch.cuda.FloatTensor)
            data, labels = data.cuda(), labels.cuda()
        
        optimizer.zero_grad()
        target = model(data)
        loss = criterion(target, torch.unsqueeze(labels, 1))
        train_loss = loss.item()
        loss.backward()
        optimizer.step()
    
    valid_loss = 0.0
    with torch.set_grad_enabled(False):
        for val_data, val_labels in bowl_vdl:
            if device.type == "cuda":
                val_data = val_data.type(torch.cuda.FloatTensor)
                val_labels = val_labels.type(torch.cuda.FloatTensor)
                val_data, val_labels = val_data.cuda(), val_labels.cuda()
                
            target = model(val_data)
            loss = criterion(target, torch.unsqueeze(val_labels, 1))
            valid_loss = loss.item()
            
    del val_data, val_labels, data, labels
    torch.cuda.empty_cache()
    #if e % 20 == 0:
    print('Epoch {} \t Training Loss: {} \t Validation Loss: {}'.format(e+1, train_loss, valid_loss))
    
    if min_valid_loss > valid_loss:
        min_valid_loss = valid_loss
#         torch.save({
#             'model_state_dict': model.state_dict(),
#             'optimizer_state_dict': optimizer.state_dict(),
#             }, os.path.join('/gdrive', 'MyDrive', 'Assignment 2', 'bowl.pth'))
```

```python
train_path = join(ISBI_, "train")
val_path = join(ISBI_, "val")

augs = A.Compose([
    A.Resize(IMG_H, IMG_W),
    A.RandomBrightnessContrast(brightness_by_max=False, p=0.5),
    A.Flip(p=0.5),
    A.ElasticTransform(p=0.5),
    A.GridDistortion(p=0.5), 
    A.Rotate(p=0.5),
], p=0.8)

train_d = ISBI_D(train_path, augs)
train_dataloader = DataLoader(train_d, batch_size=1, shuffle=True)
# img, mask = train_d.__getitem__(10)
# visualize(img.squeeze(0), mask)

val_d = ISBI_D(val_path, augs)
val_dataloader = DataLoader(val_d, batch_size=1, shuffle=True)

train_features, train_labels = next(iter(train_dataloader))
val_features, val_labels = next(iter(val_dataloader))
print("Training batch shape:", train_features.size(), train_labels.size())
print("Validation batch shape:", val_features.size(), val_labels.size())
```

```python
model = UNet(1, 1)

device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")

if device.type == "cuda":
    model.cuda()
    model = model.type(torch.cuda.FloatTensor)

criterion = torch.nn.BCEWithLogitsLoss()
optimizer = torch.optim.SGD(model.parameters(), lr=0.01)
epochs = 1000
min_valid_loss = np.inf

for e in range(epochs):
    train_loss = 0.0
    for data, labels in train_dataloader:
        if device.type == "cuda":
            data = data.type(torch.cuda.FloatTensor)
            labels = labels.type(torch.cuda.FloatTensor)
            data, labels = data.cuda(), labels.cuda()
      
        target = model(torch.unsqueeze(data, 0))
        loss = criterion(target.squeeze(1), labels.squeeze(1))
        train_loss = loss.item()
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
    
    valid_loss = 0.0
    with torch.set_grad_enabled(False):
        for val_data, val_labels in val_dataloader:
            if device.type == "cuda":
                val_data = val_data.type(torch.cuda.FloatTensor)
                val_labels = val_labels.type(torch.cuda.FloatTensor)
                val_data, val_labels = val_data.cuda(), val_labels.cuda()
                
            target = model(torch.unsqueeze(val_data, 0))
            loss = criterion(target.squeeze(1), val_labels.squeeze(1))
            valid_loss = loss.item()
            
    del val_data, val_labels, data, labels
    torch.cuda.empty_cache()
    
    if e % 20 == 0:
        print('Epoch {} \t Training Loss: {} \t Validation Loss: {}'.format(e+1, train_loss, valid_loss))
        
    if min_valid_loss > valid_loss:
      #print('Validation Loss Decreased({:.6f}--->{:.6f}) \t Saving The Model'.format(
       #   min_valid_loss, valid_loss
      #))
        min_valid_loss = valid_loss
#         torch.save(model.state_dict(), os.path.join('/gdrive', 'MyDrive', 'Assignment 2', 'train_model.pth'))
```
